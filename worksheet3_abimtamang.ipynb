Create a new column, age group, that categorizes passengers into "child" (age < 18), "adult" (age
18{64), and "senior" (age 65 and above).


```python
total_fare=df['Fare'].sum()
total_fare
```

```python
total_firstclass_fare=df[df['Pclass']==1]['Fare'].sum()
total_firstclass_fare
```

```python
fare_proporrtion_firstclass =total_firstclass_fare / total_fare
fare_proporrtion_firstclass
```

```python
total_second_fare=df[df['Pclass']==2]['Fare'].sum()
total_second_fare
```

```python
fare_proportion_secondclass =total_second_fare / total_fare
fare_proportion_secondclass
```

```python
total_thirdclass_fare=df[df['Pclass']==3]['Fare'].sum()
total_thirdclass_fare
```

```python
fare_proportion_thirdclass =total_thirdclass_fare / total_fare
fare_proportion_thirdclass
```

```python
def cat_age(age):
  if age < 18 :
    return 'child'
  elif age >= 18 and age < 64:
    return 'adult'
  else:
    return 'senior'

df['age_group']=df['Age'].apply(cat_age)
df
```

```python
total_passenger = df['PassengerId'].count()
total_passenger
```

```python
df['age_group'].value_counts()
```

```python
adult_proportion=df['age_group'].value_counts()['adult']/total_passenger
adult_proportion
```

```python
senior_proportion=df['age_group'].value_counts()['senior']/total_passenger
senior_proportion
```

```python
child_proportion=df['age_group'].value_counts()['child']/total_passenger
child_proportion
```
